```python
class DataPreprocessor:
    def __init__(self):
        self.bounds_ = {}
        self.pipeline = None

    def fit_outlier_bounds(self, X):
        # Calcul des bornes pour chaque colonne numérique
        for col in X.select_dtypes(include=[np.number]).columns:
            q1 = np.percentile(X[col], 25)
            q3 = np.percentile(X[col], 75)
            lower_bound = q1 - 1.5 * (q3 - q1)
            upper_bound = q3 + 1.5 * (q3 - q1)
            self.bounds_[col] = (lower_bound, upper_bound)
        return self

    def transform_outliers(self, X):
      # Appliquer les bornes calculées pour gérer les valeurs aberrantes
        X = X.copy()
        for col, (lower_bound, upper_bound) in self.bounds_.items():
            X.loc[X[col] < lower_bound, col] = lower_bound
            X.loc[X[col] > upper_bound, col] = upper_bound
        return X

    def replace_missing_data(self, data, method):
        # Imputer les valeurs manquantes en fonction de la méthode choisie
        for col in data.select_dtypes(include=[np.number]).columns:
            if method == "median":
                data[col].fillna(data[col].median(), inplace=True)
            elif method == "mean":
                data[col].fillna(data[col].mean(), inplace=True)
        for col in data.select_dtypes(include=['object']).columns:
            data[col].fillna(data[col].mode().iloc[0], inplace=True)

    def preprocess_data(self, X_train, X_test, missing_data_method='median'):
      # Traitement des valeurs manquantes
        self.replace_missing_data(X_train, missing_data_method)
        self.replace_missing_data(X_test, missing_data_method)

        # Traitement des valeurs aberrantes
        self.fit_outlier_bounds(X_train)
        X_train = self.transform_outliers(X_train)
        X_test = self.transform_outliers(X_test)

        # Sélection des colonnes catégorielles et numériques
        categorical_cols = X_train.select_dtypes(include='object').columns.tolist()
        numeric_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()

        # Pipeline de prétraitement avec encodage pour les variables catégorielles et standardisation pour les variables numériques
        self.pipeline = Pipeline(steps=[('preprocessor', ColumnTransformer(
                transformers=[
                    ('cat', OneHotEncoder(drop='first', sparse=False), categorical_cols),
                    ('num', StandardScaler(), numeric_cols)
                ], remainder='passthrough'
            ))
        ])

        # Transformation des ensembles d'entraînement et de test
        X_train_transformed = self.pipeline.fit_transform(X_train)
        X_test_transformed = self.pipeline.transform(X_test)

        # Convertir les résultats transformés en DataFrame avec les noms de colonnes
        X_train_transformed_df = pd.DataFrame(X_train_transformed, columns=self.pipeline.named_steps['preprocessor'].get_feature_names_out())
        X_test_transformed_df = pd.DataFrame(X_test_transformed, columns=self.pipeline.named_steps['preprocessor'].get_feature_names_out())
        return X_train_transformed_df, X_test_transformed_df
```

SubPlots

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the reusable plotting function
def create_subplots(df, features, plot_type, y=None, palette='Set2', bar_type='grouped', sample_size=None):
    """
    Create subplots for different categorical feature plots.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the data.
        features (list): List of categorical features to plot.
        plot_type (str): Type of plot ('bar', 'count', 'pie', 'strip', 'swarm').
        y (str): Target variable for strip/swarm plots (optional, must be numerical).
        palette (str): Color palette for the plots.
        bar_type (str): Bar plot type ('grouped' or 'stacked').
        sample_size (int): Number of samples for swarm plots (optional).
    """
    fig, axes = plt.subplots(nrows=(len(features) + 1) // 2, ncols=2, figsize=(24, 18))
    axes = axes.flatten()

    for i, feature in enumerate(features):
        if i < len(axes):
            if plot_type in ['swarm', 'strip'] and sample_size is not None and len(df) > sample_size:
                # Ensure sample_size does not exceed dataset size
                df_sampled = df.sample(min(sample_size, len(df)), random_state=5764)
            else:
                df_sampled = df

            if plot_type == 'bar':
                if bar_type == 'grouped':
                    sns.countplot(data=df, x=feature, ax=axes[i], palette=palette)
                elif bar_type == 'stacked':
                    if y is not None and y in df.columns:
                        crosstab = pd.crosstab(df[feature], df[y])
                        crosstab.plot(kind='bar', stacked=True, ax=axes[i], colormap=palette)
                axes[i].set_title(f'Distribution of {feature}', fontsize=16)
            elif plot_type == 'count':
                sns.countplot(data=df, x=feature, ax=axes[i], palette=palette)
                axes[i].set_title(f'Count Plot of {feature}', fontsize=16)
            elif plot_type == 'pie':
                feature_counts = df[feature].value_counts()
                axes[i].pie(feature_counts, labels=feature_counts.index, autopct='%1.1f%%',
                            startangle=90, colors=sns.color_palette(palette, len(feature_counts)))
                axes[i].set_title(f'Pie Chart of {feature}', fontsize=16)
            elif plot_type == 'strip':
                sns.stripplot(data=df_sampled, x=feature, y=y, ax=axes[i], palette=palette, jitter=True)
                axes[i].set_title(f'Strip Plot of {feature} vs {y}', fontsize=16)
            elif plot_type == 'swarm':
                sns.swarmplot(data=df_sampled, x=feature, y=y, ax=axes[i], palette=palette)
                axes[i].set_title(f'Swarm Plot of {feature} vs {y}', fontsize=16)

    # Remove extra axes
    for j in range(len(features), len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()

# Example DataFrame
df_cleaned = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Female', 'Male'],
    'Customer_Type': ['Loyal', 'Disloyal', 'Loyal', 'Disloyal'],
    'Type_of_Travel': ['Business', 'Leisure', 'Business', 'Leisure'],
    'Class': ['Economy', 'Business', 'Economy', 'First'],
    'Age': [34, 25, 45, 29],
    'satisfaction': ['Satisfied', 'Dissatisfied', 'Satisfied', 'Dissatisfied']
})

# Example categorical features
categorical_features = ['Gender', 'Customer_Type', 'Type_of_Travel', 'Class']

# Generate plots

# Grouped bar plot
create_subplots(df_cleaned, categorical_features, 'bar', palette='Set3', bar_type='grouped')

# Stacked bar plot
create_subplots(df_cleaned, categorical_features, 'bar', y='satisfaction', palette='Set3', bar_type='stacked')

# Count plot
create_subplots(df_cleaned, categorical_features, 'count', palette='Paired')

# Pie chart
create_subplots(df_cleaned, categorical_features, 'pie', palette='Pastel1')


```

Confusion Matrix

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

# Calculate the confusion matrix
cm = confusion_matrix(y_val, dt_pred, labels=dt.classes_)

# Define the labels for display
labels = dt.classes_

# Create a custom colormap with blue, yellow, and green
from matplotlib.colors import ListedColormap
cmap = ListedColormap(["blue", "yellow", "green"])

# Plot the confusion matrix with the custom colormap
plt.figure(figsize=(8, 6))
plt.imshow(cm, interpolation='nearest', cmap=cmap)
plt.colorbar()
plt.title("Confusion Matrix")

# Add axis labels and tick marks
tick_marks = np.arange(len(labels))
plt.xticks(tick_marks, labels, rotation=45)
plt.yticks(tick_marks, labels)

# Add text annotations
thresh = cm.max() / 2.  # Threshold for text color
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, format(cm[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import GridSearchCV
import pandas as pd
import matplotlib.pyplot as plt

# Define the model
xgb = XGBClassifier(objective='binary:logistic', random_state=31)

# Define hyperparameter grid
xgb_params = {
    'max_depth': [4, 6],
    'min_child_weight': [3, 5],
    'learning_rate': [0.2, 0.3],
    'n_estimators': [15],
    'subsample': [0.7],
    'colsample_bytree': [0.7]
}

# Define scoring metrics
scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1'
}

# GridSearchCV for hyperparameter tuning
xgb_cv = GridSearchCV(estimator=xgb, param_grid=xgb_params, scoring=scoring, refit='f1', cv=5, verbose=1, n_jobs=-1)

# Fit the model
xgb_cv.fit(X_tr, y_tr)

# Predictions on validation set
xgb_pred = xgb_cv.predict(X_val)

# Confusion Matrix
cm = confusion_matrix(y_val, xgb_pred, labels=xgb_cv.best_estimator_.classes_)

# Create a DataFrame for evaluation metrics
xgb_results = pd.DataFrame({
    'model': ['XGBoost'],
    'accuracy': [accuracy_score(y_val, xgb_pred)],
    'precision': [precision_score(y_val, xgb_pred)],
    'recall': [recall_score(y_val, xgb_pred)],
    'f1': [f1_score(y_val, xgb_pred)]
})

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=xgb_cv.best_estimator_.classes_)
disp.plot(values_format='')
plt.show()

# Combine results into an existing results DataFrame
# Assuming `results` is an existing DataFrame to which you are appending
try:
    results = pd.concat([results, xgb_results])
except NameError:
    # If `results` doesn't exist, create it
    results = xgb_results

# Display the updated results DataFrame
print(results)
```
